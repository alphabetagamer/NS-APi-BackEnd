```python
filename = "products.js"
c=[]
# open the file for reading
filehandle = open(filename, 'r')
while True:
    # read a single line
    line = filehandle.readline()
    c.append(line)
    if not line:
        break
    print(line)

# close the pointer to that file
filehandle.close()
```

```python
c=[i for i in c if "//" not in i]
```
